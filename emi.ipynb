## Exploratory Data Analysis

```python
# Required libraries: pandas, numpy, matplotlib, seaborn, scipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# assume `df` is already loaded and has the columns from previous chats

# ---- basic cleaning used in analysis ----
df = df.copy()
df.columns = df.columns.str.strip()

# normalize target
df['emi_eligibility'] = df['emi_eligibility'].astype(str).str.strip().str.lower()
df['target_approved'] = df['emi_eligibility'].apply(lambda x: 1 if 'eligible' in x else 0)

# ensure numeric financial columns are numeric
financial_cols = [
    "monthly_salary", "bank_balance", "credit_score", "current_emi_amount",
    "requested_amount", "max_monthly_emi", "emergency_fund"
]
for c in financial_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ---- 1) EMI eligibility distribution across emi_scenario ----
scenario_col = "emi_scenario"
emi_summary = df.groupby([scenario_col, 'emi_eligibility']).size().reset_index(name='count')
emi_summary['total_in_scenario'] = emi_summary.groupby(scenario_col)['count'].transform('sum')
emi_summary['pct_in_scenario'] = (emi_summary['count'] / emi_summary['total_in_scenario'] * 100).round(2)

# pivot for plotting and inspection
emi_pivot_pct = emi_summary.pivot(index=scenario_col, columns='emi_eligibility', values='pct_in_scenario').fillna(0)
emi_pivot_count = emi_summary.pivot(index=scenario_col, columns='emi_eligibility', values='count').fillna(0)

print("Top 10 scenarios by volume:")
print(df[scenario_col].value_counts().head(10))

# stacked percent bar
plt.figure(figsize=(12,6))
emi_pivot_pct.sort_values(by=('Eligible' if 'Eligible' in emi_pivot_pct.columns else emi_pivot_pct.columns[0]),
                          ascending=False).plot(kind='bar', stacked=True, legend=True)
plt.title('EMI eligibility composition (%) by scenario')
plt.xlabel('EMI scenario')
plt.ylabel('Percent within scenario')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# approval rate per scenario
approval_rate_by_scenario = (df.assign(approved=df['target_approved'])
                              .groupby(scenario_col)['approved']
                              .mean()
                              .mul(100)
                              .round(2)
                              .reset_index(name='approval_rate_pct'))
print("Approval rate by scenario (top 20 by volume):")
print(approval_rate_by_scenario.merge(df[scenario_col].value_counts().rename('volume'),
                                      left_on=scenario_col, right_index=True)
                                  .sort_values(['volume'], ascending=False).head(20))

plt.figure(figsize=(10,5))
sns.barplot(data=approval_rate_by_scenario.sort_values('approval_rate_pct', ascending=False).head(15),
            x='approval_rate_pct', y=scenario_col)
plt.xlabel('Approval rate (%)')
plt.ylabel('EMI scenario')
plt.title('Top 15 scenarios by approval rate')
plt.tight_layout()
plt.show()

# ---- 2) Correlation: financial variables vs loan approval ----
# choose available financial vars
available_financial = [c for c in financial_cols if c in df.columns]
corr_results = []
for c in available_financial:
    subset = df[[c, 'target_approved']].dropna()
    if subset.shape[0] < 30:
        continue
    # use point-biserial (equivalent to Pearson with binary target)
    try:
        r, p = stats.pointbiserialr(subset['target_approved'], subset[c])
    except Exception:
        r, p = stats.pearsonr(subset[c], subset['target_approved'])
    corr_results.append({'variable': c, 'r': float(r), 'p_value': float(p), 'n': int(subset.shape[0])})

corr_df = pd.DataFrame(corr_results).sort_values(by='r', key=lambda s: s.abs(), ascending=False)
print("Point-biserial correlations (financial vars vs approval):")
print(corr_df)

# Spearman correlation matrix across financial vars + target
corr_matrix = df[available_financial + ['target_approved']].corr(method='spearman').round(3)
print("Spearman correlation snippet:")
print(corr_matrix.loc[:, ['target_approved']].sort_values('target_approved', ascending=False))

# scatter/box visuals for top correlated variables
top_vars = corr_df['variable'].tolist()[:4]
for v in top_vars:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='target_approved', y=v, data=df, showfliers=False)
    plt.title(f'{v} by approval (0=Rejected,1=Approved)')
    plt.tight_layout()
    plt.show()

# ---- 3) Demographic patterns and risk factor relationships ----
# Gender vs approval
if 'gender' in df.columns:
    gender_tab = pd.crosstab(df['gender'].fillna('Missing'), df['target_approved'])
    gender_rates = (gender_tab[1] / gender_tab.sum(axis=1) * 100).round(2)
    print("Approval rates by gender:")
    print(gender_rates.sort_values(ascending=False))
    # chi-square
    try:
        chi2, p, dof, ex = stats.chi2_contingency(gender_tab)
        print("Gender vs approval chi2 p-value:", p)
    except Exception:
        pass
    # bar plot
    plt.figure(figsize=(6,4))
    gender_rates.sort_values(ascending=False).plot(kind='bar')
    plt.ylabel('Approval rate (%)')
    plt.title('Approval rate by gender')
    plt.tight_layout()
    plt.show()

# Age patterns: summary and binned rates
if 'age' in df.columns:
    df['age_bin'] = pd.cut(df['age'], bins=[15,25,35,45,55,65,100], right=False)
    age_rates = (df.groupby('age_bin')['target_approved'].mean().mul(100)).round(2)
    print("Approval rate by age bin:")
    print(age_rates)
    age_counts = df['age_bin'].value_counts().sort_index()
    fig, ax1 = plt.subplots(figsize=(8,4))
    age_rates.plot(kind='bar', ax=ax1)
    ax1.set_ylabel('Approval rate (%)')
    plt.title('Approval rate by age bin')
    plt.tight_layout()
    plt.show()

# Marital status vs approval
if 'marital_status' in df.columns:
    ms_tab = pd.crosstab(df['marital_status'].fillna('Missing'), df['target_approved'])
    ms_rates = (ms_tab[1] / ms_tab.sum(axis=1) * 100).round(2)
    print("Approval rates by marital status (top categories):")
    print(ms_rates.sort_values(ascending=False).head(10))

# Existing loans and current EMI as risk signals
if 'existing_loans' in df.columns:
    el_tab = pd.crosstab(df['existing_loans'].fillna('Missing'), df['target_approved'])
    el_rates = (el_tab[1] / el_tab.sum(axis=1) * 100).round(2)
    print("Approval rate by existing_loans:")
    print(el_rates)
if 'current_emi_amount' in df.columns:
    medians = df.groupby('target_approved')['current_emi_amount'].median().rename({0:'rejected_median',1:'approved_median'})
    print("Median current EMI by approval:", medians.to_dict())

# Combine risk factors: high requested_amount & low credit_score
if set(['requested_amount','credit_score']).issubset(df.columns):
    df['high_requested_low_credit'] = ((df['requested_amount'] > df['requested_amount'].quantile(0.75)) &
                                       (df['credit_score'] < df['credit_score'].quantile(0.25)))
    risk_cross = pd.crosstab(df['high_requested_low_credit'], df['target_approved'])
    print("High requested & low credit cross-tab:")
    print(risk_cross)

# ---- 4) Statistical summaries & business insights ----
summary = {}
summary['rows'] = df.shape[0]
summary['overall_approval_rate_pct'] = float(df['target_approved'].mean() * 100)
summary['top_scenarios_by_volume'] = df[scenario_col].value_counts().head(10).to_dict()
summary['lowest_approval_scenarios'] = approval_rate_by_scenario.sort_values('approval_rate_pct').head(5).to_dict(orient='records')
summary['highest_approval_scenarios'] = approval_rate_by_scenario.sort_values('approval_rate_pct', ascending=False).head(5).to_dict(orient='records')

print("Summary business insights:")
print(summary)

# quick table: scenarios with high volume but low approval (actionable)
merged = approval_rate_by_scenario.merge(df[scenario_col].value_counts().rename('volume'), left_on=scenario_col, right_index=True)
actionable = merged[(merged['volume'] >= merged['volume'].quantile(0.75)) & (merged['approval_rate_pct'] <= merged['approval_rate_pct'].quantile(0.25))]
print("High-volume, low-approval scenarios (investigate):")
print(actionable.sort_values(['approval_rate_pct']).head(10))
```
